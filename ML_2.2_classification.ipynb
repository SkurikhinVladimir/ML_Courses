```python
df = load_iris(as_frame=True).frame
```

```python
df['target_virginica'] = df['target'].apply(lambda x: 0 if x != 2 else 1)
```

```python
display(df['target'].value_counts())
display(df['target_virginica'].value_counts())
```

```python
features = df.drop(['target_virginica','target'], axis = 1)
target = df['target_virginica']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred = clf.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
print(f"Accuracy score {accuracy_score(y_test, y_pred)}")
```

```python
class MyLogisticRegression:
    def __init__(self, step_size=0.9, iterations=1000):
        self.step_size = step_size
        self.iterations = iterations

    @staticmethod
    def sigmoid(x):
        return 1 / (1 + np.exp(-x))

    def fit(self, train_features, train_target):
        X = np.concatenate((np.ones((train_features.shape[0], 1)), train_features), axis=1)
        y = train_target
        w = np.zeros(X.shape[1])

        for i in range(self.iterations):
            gradient = X.T @ (self.sigmoid(X @ w)-y) / X.shape[0]
            w -= self.step_size * gradient

        self.w = w[1:]
        self.w0 = w[0]

    def predict(self, test_features):
        return self.sigmoid(test_features.dot(self.w) + self.w0) > 0.5
```

```python
clf = MyLogisticRegression()
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
print(f"Accuracy score {accuracy_score(y_test, y_pred)}")
```
